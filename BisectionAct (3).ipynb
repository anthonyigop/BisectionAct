```python
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
    return 3*x**5 - 6*x**2 + 11*x - 6

def f2(x):
    return np.exp(x) + 3*x**3

def f3(x):
    return np.sin(x) + np.exp(x) - 3*x

def f4(x):
    return (4*x - 3*x**4) / np.exp(np.sin(x))

def f5(x):
    return (2 + np.cos(x)) / (3 * np.exp(np.sin(x)))

def bisection_method(f, a, b, tol=0.05, max_iter=100):
    if f(a) * f(b) > 0:
        return None  

    for i in range(max_iter):
        mid = (a + b) / 2
        f_mid = f(mid)

        if abs(f_mid) < tol or abs(b - a) < tol:
            return mid, i+1  

        if f(a) * f_mid < 0:
            b = mid  
        else:
            a = mid  

    return None  

def plot_function(f, a, b, title, root=None):
    x = np.linspace(a, b, 1000)
    y = f(x)
    
    plt.figure(figsize=(6, 4))
    plt.plot(x, y, label=title, color='blue')
    plt.axhline(0, color='black', linewidth=0.8)
    
    if root is not None:
        plt.scatter(root, f(root), color='red', zorder=5, label="Root")
    
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

functions = [
    (f1, -1, 2, "Equation 1: 3x^5 - 6x^2 + 11x - 6"),
    (f2, -1, 1, "Equation 2: e^x + 3x^3"),
    (f3, -2, 2, "Equation 3: sin(x) + e^x - 3x"),
    (f4, -1, 2, "Equation 4: (4x - 3x^4)/e^sin(x)"),
    (f5, -1, 1, "Equation 5: (2 + cos(x))/(3e^sin(x))")
]

for f, a, b, title in functions:
    print("-" * 100)
    print(f"Processing: {title}")
    print(f"Initial interval: a = {a}, b = {b}")

    root_bisection = bisection_method(f, a, b)

    if root_bisection:
        root, iterations = root_bisection
        print(f"Root found: {root:.5f} (Bisection Method) in {iterations} iterations.")
        plot_function(f, a, b, title, root=root)
    else:
        print(f"No real root found using Bisection Method. Possible reasons:")
        print(" - The function may have complex (imaginary) roots.")
        print(" - The function does not change sign in the given interval.")
        plot_function(f, a, b, title)

    print(" " * 100)  


#
```

```python
# Reflection and Discussion  Bisection Method

## 1. How did the graph help in identifying the best initial guesses for the bisection method?  
The graph provides a visual representation of the functions behavior, helping to identify intervals where the function crosses the x-axis. These crossing points indicate possible root locations. By observing where the function changes sign (f(a) and f(b) have opposite signs), we can select the best initial guesses for applying the bisection method effectively.  

## 2. What is the role of the tolerance  in determining the accuracy of the root?  
The tolerance  determines the stopping condition for the method.  
- Smaller tolerance  Higher accuracy but requires more iterations.  
- Larger tolerance  Lower accuracy but requires fewer iterations.  
This balance between accuracy and efficiency is crucial when choosing an appropriate tolerance value.  

## 3. How does the number of iterations change with different initial guesses?  
The number of iterations depends on the width of the initial interval [a, b]:  
- Larger interval  Takes more iterations to narrow down to the root.  
- Smaller interval  Converges faster with fewer iterations.  
Additionally, if the chosen interval does not contain a sign change, the bisection method fails to find a root.  

## 4. Which equation required fewer iterations to find the root, and why?  
The equation that required fewer iterations had a well-chosen initial interval where:  
- The function changed sign within the interval.  
- The function had a steeper slope near the root.  
Functions with sharp changes near the root converge faster, whereas functions with flatter curves take more iterations because the function values change gradually.
```
